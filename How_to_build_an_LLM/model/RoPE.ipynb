# Rotary Positional Embeddings (RoPE)

[Paper](https://arxiv.org/abs/2104.09864)

- The "absolute positional embeddings" was the Positional Encoding in the Transformer architecture. So instead of adding the positional encoding at the beginning (i.e., to the Output Embedding as shown in the Transformer architecture), we add it to the **Query** and **Key** before the **GQA** layer.
