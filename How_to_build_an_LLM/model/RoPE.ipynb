# Rotary Positional Embeddings (RoPE)

![RoPE Diagram](../showcase_images/RoPE.png)

**From Llama 1 & 2 paper:** "Rotary Embeddings [GPTNeo]. We remove the absolute positional embeddings, and instead, add rotary positional embeddings (RoPE), introduced by [Su et al. (2021)](https://arxiv.org/abs/2104.09864), at each layer of the network."
- The "absolute positional embeddings" was the Positional Encoding in the Transformer architecture.

**From Llama 3 paper:** All the Llama 3 models used $\text{Positional Embeddings} = RoPE (\theta =500{,}000)$ from Table 3 Overview of the key hyperparameters of Llama 3.


**RoPE:**

- RoPE rotates the **Query** and **Key** tensors based on their position in the sequence.
