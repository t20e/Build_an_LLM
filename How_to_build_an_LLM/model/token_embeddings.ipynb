**Embeddings** are numerical representations of words in a **higher-dimensional space**, that capture semantic, syntactic, and contextual information. Words with similar meanings are positioned close to each other. The distance between vectors encodes the **degree of similarity between words**.

A good [Embedding Space Visualization](https://projector.tensorflow.org/) by tensorflow.org |  another by [google](https://developers.google.com/machine-learning/crash-course/embeddings/embedding-space)

- The embedding layer is a **learnable lookup table**. It learns during pre-training, where its weights are updated during the optimizer via backpropagation. It becomes static during inference.

```python
import EasyJupyter
import torch
import torch.nn as nn


class TokenEmbeddings(nn.Module):
    def __init__(self, device, d_model, vocab_size):
        """Implements the Token Embeddings layer"""
        super().__init__()
        self.look_up_table = nn.Embedding(
            num_embeddings=vocab_size, embedding_dim=d_model, device=device
        )

    def forward(self, token_ids):
        return self.look_up_table(token_ids)
```

```python
# @i-c
# TEST
from llama_configs import Llama3_scaled_down

d_model = 512
vocab_size = 32000
context_len = 256
batch_size = 8
cfg = Llama3_scaled_down()

em = TokenEmbeddings(cfg.device, d_model, vocab_size)

dummy_input = torch.randint(0, vocab_size, (batch_size, context_len), device=cfg.device)
print(f"Input shape: {dummy_input.shape}")

output = em(dummy_input)
print(f"Output shape: {output.shape}")

expected_shape = (batch_size, context_len, d_model)
assert (
    output.shape == expected_shape
), f"Output shape: {output.shape} != Expected shape: {expected_shape}"
```
