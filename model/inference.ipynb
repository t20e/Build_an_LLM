# Inference

**Inference:**
- During inference (Autoregressive Generation), 
    - Steps:
      1. Feed the model a prompt
      2. The model generates Token 1
      3. Append Token 1 to a new sequence
      4. Feed that new sequence to the model
      5. It generates Token 2
      6. Append Token 2 to the sequence
      7. And repeat until the model generates an end-of-doc/end-of-text token, or we reach the maximum sequence length to stop infinite generation.
