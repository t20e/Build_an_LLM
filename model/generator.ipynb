# Generator

![generator](../showcase_images/generator.png)

The generator is the final `RMSNorm`, `Linear`, and `Softmax` layers to convert the Decoder's vector output back into vocabulary probabilities.

```python
import easyjupyter
import torch.nn as nn
from model.RMSNorm import RMSNorm
```

```python
from typing import TYPE_CHECKING
if TYPE_CHECKING:
    from llama_configs import BaseConfig
```

```python
class Generator(nn.Module):
    def __init__(self, cfg: BaseConfig):
        super().__init__()
        self.proj = nn.Linear(cfg.d_model, cfg.vocab_size)
        self.rms_norm = RMSNorm(cfg)

    def forward(self, x):
        # 1. Apply RMS Norm
        x = self.rms_norm(x)

        # 2. Apply the Linear layer
        # NOTE: I use nn.CrossEntropyLoss() in the loss function, which will automatically apply softmax. However, you will have to apply softmax during inference, after calling the generator!
        return self.proj(x)
```
